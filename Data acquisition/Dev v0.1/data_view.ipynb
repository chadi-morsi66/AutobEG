```python
df = df2.groupby(['brand', 'model', 'year', 'fuel_type', 'transmission', 'body_type', 'engine_capacity'])
```

```python
df.isnull().sum()
```

```python
cols = ["brand", "model", "year", "fuel_type", "transmission", "body_type"]

check = (
    df
    .groupby(cols)["engine_capacity"]
    .agg([
        lambda x: x.isna().any(),        # has_null
        lambda x: x.notna().any()        # has_non_null
    ])
)

check.columns = ["has_null", "has_non_null"]

# combinations that have BOTH
problematic = check[(check["has_null"]) & (check["has_non_null"])]

problematic
```

```python
## Filling null engines with median

cols = ["brand", "model", "year", "fuel_type", "transmission", "body_type"]

df["engine_capacity"] = (
    df.groupby(cols)["engine_capacity"]
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
df = df.drop_duplicates(['brand', 'model', 'year', 'fuel_type', 'transmission', 'body_type', 'engine_capacity'])
```

```python
cols = ["brand", "model", "year", "fuel_type", "transmission", "body_type"]

validation = (
    df
    .groupby(cols)["engine_capacity"]
    .agg(has_null=lambda x: x.isna().any(),
         has_non_null=lambda x: x.notna().any())
)

# combinations that still have both
problematic = validation[
    (validation["has_null"]) & (validation["has_non_null"])
]

problematic
```

```python
## Converting year column to integer
df["year"] = df["year"].replace([np.inf, -np.inf], np.nan).astype("Int64")
```

```python
df['year'].dtype
```

```python
cols = ["brand", "model", "year", "fuel_type",
        "transmission", "body_type", "engine_capacity"]

df["vehicle_id"] = (
    df[cols]
    .astype(str)
    .agg("_".join, axis=1)
)

df = df.set_index("vehicle_id")
```

## Body type standardization

```python

df.loc[
    df["body_type"].str.contains(
        r"small city car|\.minimize|capitalize",
        case=False,
        na=False
    ),
    "body_type"
] = "Hatchback"

# 4X4
df.loc[
    df["body_type"].str.contains(r"4x4|4wd|awd|off.?road", case=False, na=False),
    "body_type"
] = "4X4"

# Sports / Coupe
df.loc[
    df["body_type"].str.contains(r"sport|coupe", case=False, na=False),
    "body_type"
] = "Sports / Coupe"

# Sedan
df.loc[
    df["body_type"].str.contains(r"sedan|saloon", case=False, na=False),
    "body_type"
] = "Sedan"

# Van / Bus
df.loc[
    df["body_type"].str.contains(r"van|bus|minibus|mpv", case=False, na=False),
    "body_type"
] = "Van / Bus"

df.loc[df["body_type"] == "4X4", "body_type"] = "SUV"

df = df[df["body_type"] != "Other"]
```
