```python
import requests
from bs4 import BeautifulSoup
import re
from smolagents import Tool

TAVILY_URL = "https://api.tavily.com/search"
TAVILY_KEY = "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" 

# -------------------------
# Helper function
# -------------------------
def extract_full_article(url):
    """Extract readable text from a webpage (HTML → clean text)."""
    try:
        html = requests.get(url, timeout=20).text
        soup = BeautifulSoup(html, "html.parser")

        # remove unwanted tags
        for tag in soup(["script", "style", "noscript", "header", "footer", "nav"]):
            tag.extract()

        # extract visible text
        text = soup.get_text(" ", strip=True)
        text = re.sub(r"\s+", " ", text)

        # limit to prevent LLM overload
        return text[:15000]

    except Exception as e:
        return f"[Error scraping {url}: {e}]"


# -------------------------
# SmolAgent Tool
# -------------------------
class TavilyQueryScraper(Tool):
    name = "tavily_query_scraper"
    description = (
        "Search the web using Tavily for a natural language query, fetch the "
        "first 10 URLs, scrape full article text, and return merged content."
    )
    inputs = {
        "query": {
            "type": "string",
            "description": "Search query like 'price of apple today'"
        }
    }
    output_type = "string"

    def forward(self, query):
        """Search Tavily → scrape top URLs → return clean merged text."""
        payload = {
            "api_key": TAVILY_KEY,
            "query": query,
            "search_depth": "basic",
            "include_answer": False,
            "max_results": 10
        }

        # 1. Perform Tavily search
        try:
            response = requests.post(TAVILY_URL, json=payload, timeout=20)
            data = response.json()
        except Exception as e:
            return f"[ERROR contacting Tavily: {e}]"

        results = data.get("results", [])
        if not results:
            return "No search results found."

        # 2. Scrape articles
        full_context = ""
        for r in results[:10]:
            url = r.get("url", "")
            if not url:
                continue

            article_text = extract_full_article(url)

            full_context += (
                f"\n\n===============================\n"
                f"Source: {url}\n"
                f"-------------------------------\n"
                f"{article_text}\n"
            )

        return full_context.strip()
```

```python
import requests
from smolagents import Tool

TAVILY_URL = "https://api.tavily.com/search"
TAVILY_KEY = "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" 


class TavilySnippetSearch(Tool):
    name = "tavily_snippet_search"
    description = (
        "Search the web using Tavily for a natural language query and return "
        "only the top result snippets (no scraping). Useful for fast queries."
    )

    inputs = {
        "query": {
            "type": "string",
            "description": "Search query such as 'price of apple today'"
        }
    }

    output_type = "string"

    def forward(self, query):
        """Search Tavily and return only concise snippets (no URL scraping)."""
        payload = {
            "api_key": TAVILY_KEY,
            "query": query,
            "search_depth": "basic",
            "include_answer": True,
            "max_results": 10
        }

        # 1. Tavily API call
        try:
            response = requests.post(TAVILY_URL, json=payload, timeout=15)
            data = response.json()
        except Exception as e:
            return f"[ERROR contacting Tavily: {e}]"

        results = data.get("results", [])
        answer = data.get("answer", "")

        if not results:
            return "No snippets available."

        # 2. Build clean text from snippets
        output = []

        if answer:
            output.append(f"AI-Generated Answer:\n{answer}\n")

        for r in results[:10]:
            title = r.get("title", "")
            snippet = r.get("content", "")
            url = r.get("url", "")

            output.append(
                f"\n=== Result ===\n"
                f"Title: {title}\n"
                f"URL: {url}\n"
                f"Snippet: {snippet}\n"
            )

        return "\n".join(output).strip()
```
